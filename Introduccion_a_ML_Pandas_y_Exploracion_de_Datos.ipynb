### ¿Qué es Pandas?

##### Pandas es una librería ampliamente utilizada de código abierto dentro de los desarrolladores de Python, y se utiliza sobre todo dentro del ámbito de Data Science y Machine Learning. Su principal ventaja está dada por la existencia de dos objetos que permite manipular los datos de forma tabular convenientemente: Series y DataFrames
##### Series:
> Una Series es un objeto similar a un vector unidimensional. Contiene un array de valores y un array de etiquetas asociados a estos valores denominado índice. <br>
> Cuando no especificamos un índice para los datos, se asigna por defecto un índice formado por valores enteros de 0 a N-1, donde N es la cantidad de valores en la serie. <br>
> Los valores de la serie pueden ser de cualquier tipo de datos, pero todos los valores de una serie deben coincidir en su tipo. Las etiquetas, además de numéricas, también pueden ser de tipo cadena de caracteres. Al igual que los arrays de NumPy, permiten pasar una lista de elementos (índices) para seleccionar un subconjunto de valores. <br>
##### DataFrames:
> Son una estructura de datos tabular que contiene una colección de columnas, cada una de las cuales tiene un tipo de datos determinado (number, string, boolean, etc.). <br>
> Podemos pensar un objeto DataFrame como un diccionario de Series "alineadas" (que comparten el mismo índice). <br>
> Se pueden indexar a través del índice y el nombre de la columna, o al igual que en un array de numpy, utilizando los valores de posición.

##### Doc: _https://pandas.pydata.org/docs/_ <br> _https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html_ <br> _https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.html_

### Imputación de valores nulos:
##### El tratamiento de los valores faltantes en un dataset es muy importante, ya que es muy habitual encontrarnos con ellos cuando trabajamos con datasets reales.
##### Como regla general, hay dos caminos a tomar frente a la presencia de valores nulos:
+ Imputar los valores faltantes
+ Eliminar la fila/columna que los contiene <br>
##### Ambos procesos tienen sus ventajas y desventajas, en el primer caso estamos frente a una situación en la que podríamos perder la representabilidad real de los datos, o perdiendo la variabilidad que es de la que se nutren los modelos; en el segundo caso podríamos estar perdiendo información que puede ser de interés.
##### No hay método simple para tratar este problema, y la mejor opción es la de hacer una explroación exhaustiva de los datos, compararlos y corroborar a cada estapa que no se esté alterando la información original de forma sustancial (como cambiar la forma de la distribución en una variable continua)

```python
# Seteo para mostrar mas filas y columnas
pd.set_option('display.max_columns', 250)
pd.set_option('display.max_rows', 250)
```

```python
# Identificamos los valores nulos utilizando el método .isnull()
df_temp.isnull().sum()
```

+ Podemos ver que hay una gran cantidad de nulos en las columnas de valores
+ Vamos a explorar el método fillna
https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html

```python
# Con el método fillna podemos imputar los valores de una columna, con un método propio de la funcióin o asignando un valor nosotros mismo
# Tomemos de ejemplo la columna 'Y1994'
df_temp['Y1994'].fillna(method = 'backfill') # Asigna el siguiente valor válido
```

```python
# Superponemos los histogramas de los valores originales sobre los valores con la imputación
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(13,7))

df_temp['Y1994'].fillna(method = 'backfill').hist(bins=40)

df_temp['Y1994'].hist(bins=40)
```

+ Podemos observar que los valores imputados producen unadistribución muy similar a la original
+ Sin embargo, como el dataset contiene valores de temperaturas para distintos paìses, podrían estar asignandose valores desproporcionados (por ej, si una temperatura faltante para 'Brasil' es imputada con el siguiente valor el cual corresponde a 'Finlandia'). Es por esto que debemos tener cuidado en la imputación de valores nulos.

```python
# Imputamos ahora por un valor fijo, en este caso el valor promedio de la columna (puede ser otro)

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(13,7))

df_temp['Y1994'].fillna(value = df_temp['Y1994'].mean()).hist(bins=40) # Asigna el siguiente valor válido

df_temp['Y1994'].hist(bins=40)
```

+ En este caso podemos ver que la distribución tiene a ser mas concentrada en la media (menor varianza), esto puede generar una gran pérdida en la variabilidad de los datos

##### Imputaciones mas sofisticadas pueden realizarse si se conocen mejor los datos.
##### Por ejemplo en este caso podría imputarse para cada valor faltante de cada país, el promedio entre el valor anterior y posterior para ése país, habiendo previamente ordenado los valores, de tal manera que los meses queden en el orden correcto (si el valor faltante se encuentra para el mes de abril, el promedio sería entre el valor de marzo y mayo). Ésto sería mas consistente que imputar simplemente por la media.
